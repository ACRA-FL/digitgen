```python
from digitgen import DigitGenerator
from digitgen.augmentation import RandomImageWidthChange,AscendingImageWidthChange,GaussianNoise,SPNoise, SpeckleNoise, PoissonNoise
from matplotlib import pyplot as plt
```

## Basic UseCase

```python
digit_gen = DigitGenerator(8,samples=1000,image_size=(512, 32))
digit_gen.generate_digits()
digit_gen.generate_digit_config()
ret_arr, ret_ann = digit_gen.generate_dataset()
```

```python
plt.imshow(ret_arr[0])
```

## GrayScale Images

```python
digit_gen = DigitGenerator(8,samples=1000,image_size=(512, 32),gray_scale=True)
digit_gen.generate_digits()
digit_gen.generate_digit_config()
```

```python
ret_arr, ret_ann = digit_gen.generate_dataset()
```

```python
plt.imshow(ret_arr[0])
```

## Generate Digits with choice probability

Input the probability of each character

```python
digit_gen = DigitGenerator(8,samples=1000,image_size=(512, 32),gray_scale=True,transpose=True)
digit_gen.generate_digits_with_choice_probability({"0": 0.5, "1": 0.1, "2": 0.2, "3": 0.2})
digit_gen.generate_digit_config()
```

## Generate Digits with Positional Probability

Input the probability of each position check below example

```python
digit_gen = DigitGenerator(8,samples=1000,image_size=(512, 32),gray_scale=True,transpose=True)
digit_gen.generate_digits_with_position_probability({
    0: {"0": 0.5,
        "1": 0.1,
        "2": 0.2,
        "3": 0.2},
    1: {"4": 1},
    2: {"5": 0.5,
        "8": 0.5},
    "default": {
        "0": 0.1,
        "1": 0.1,
        "2": 0.1,
        "3": 0.1,
        "4": 0.1,
        "5": 0.1,
        "6": 0.1,
        "7": 0.1,
        "8": 0.1,
        "9": 0.1}
})
digit_gen.generate_digit_config()
```

## Random Spaces

```python
digit_gen = DigitGenerator(10, samples=100, image_size=(128, 48))
digit_gen.generate_digits()
digit_gen.generate_spaces(space_type="random", num_spaces=3)
digit_gen.generate_digit_config()
ret_arr, ret_ann = digit_gen.generate_dataset()
```

```python
plt.imshow(ret_arr[0])
```

## Space Dividing Into Segments

```python
digit_gen = DigitGenerator(12, samples=100, image_size=(128, 48))
digit_gen.generate_digits()
digit_gen.generate_spaces(space_type="space", sectors=3, spaces_per_sector=2)
digit_gen.generate_digit_config()
ret_arr, ret_ann = digit_gen.generate_dataset()
```

```python
plt.imshow(ret_arr[0])
```

## Changing Spaces Between

```python
digit_gen = DigitGenerator(10, samples=20, image_size=(512, 48))
digit_gen.generate_digits()
digit_gen.generate_digit_config()
digit_gen.add_width_digits(pixel_value=5)
ret_arr, ret_ann = digit_gen.generate_dataset()
```

```python
plt.imshow(ret_arr[0])
```

## Changing Spaces Between Characters Random

```python
digit_gen = DigitGenerator(10, samples=100, image_size=(512, 48))
digit_gen.generate_digits()
digit_gen.generate_digit_config()
digit_gen.add_width_digits(mode="random", pixel_range=[0, 10])
ret_arr, ret_ann = digit_gen.generate_dataset()
```

```python
plt.imshow(ret_arr[0])
```

## Random Character Width

```python
digit_gen = DigitGenerator(8,samples=1000,image_size=(512, 48),augmentations =[RandomImageWidthChange(width_range=[0, 20])])
digit_gen.generate_digits()
digit_gen.generate_digit_config()
digit_gen.add_width_digits(mode="random", pixel_range=[0, 10])
ret_arr, ret_ann = digit_gen.generate_dataset()
```

```python
plt.imshow(ret_arr[0])
```

## Ascending Character size

```python
digit_gen = DigitGenerator(8,samples=1000,image_size=(512, 48),augmentations =[AscendingImageWidthChange(width_increment=5)])
digit_gen.generate_digits()
digit_gen.generate_digit_config()
digit_gen.add_width_digits(mode="random", pixel_range=[0, 10])
ret_arr, ret_ann = digit_gen.generate_dataset()
```

```python
plt.imshow(ret_arr[0])
```

## Adding Gaussian Noise

```python
digit_gen = DigitGenerator(8,samples=1000,image_size=(512, 48),gray_scale=False,augmentations =[GaussianNoise()])
digit_gen.generate_digits()
digit_gen.generate_digit_config()
digit_gen.add_width_digits(mode="random", pixel_range=[0, 10])
ret_arr, ret_ann = digit_gen.generate_dataset()
```

```python
plt.imshow(ret_arr[0])
```

## Adding SP Noise And Speckle Noise together

```python
digit_gen = DigitGenerator(8,samples=1000,image_size=(512, 48),gray_scale=False,augmentations =[SPNoise(s_vs_p=0.7),SpeckleNoise(probability=0.8)])
digit_gen.generate_digits()
digit_gen.generate_digit_config()
digit_gen.add_width_digits(mode="random", pixel_range=[0, 10])
ret_arr, ret_ann = digit_gen.generate_dataset()
```

```python
plt.imshow(ret_arr[0])
```

## Adding Poisson Noise with Random Character width

```python
digit_gen = DigitGenerator(8,samples=1000,image_size=(512, 48),augmentations =[RandomImageWidthChange(width_range=[0, 20]),PoissonNoise()])
digit_gen.generate_digits()
digit_gen.generate_digit_config()
digit_gen.add_width_digits(mode="random", pixel_range=[0, 10])
ret_arr, ret_ann = digit_gen.generate_dataset()
```

```python
plt.imshow(ret_arr[0])
```
